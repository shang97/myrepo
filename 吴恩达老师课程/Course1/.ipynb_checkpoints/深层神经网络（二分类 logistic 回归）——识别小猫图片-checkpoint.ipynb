**模型构建步骤**
- 1.初始化网络参数
- 2.前向传播  
 - 2.1 计算一层的中线性求和的部分  
 - 2.2 计算激活函数的部分（ReLU使用 $L-1$ 次，Sigmod使用1次）  
 - 2.3 结合线性求和与激活函数
- 3.计算误差
- 4.反向传播  
 - 4.1 线性部分的反向传播公式  
 - 4.2 激活函数部分的反向传播公式  
 - 4.3 结合线性部分与激活函数的反向传播公式
- 5.更新参数  

对于每个前向函数，都有一个相应的后向函数。 因此在前向传播模块的每一步都会 **cache** 一些值，cache的值对计算梯度很有用，在反向传播模块中，将使用 cache 来计算梯度

### 模型构建
#### 初始化网络参数
##### 初始化两层神经网络的参数
对于一个两层的神经网络结构而言，模型结构是线性->ReLU->线性->sigmod函数

```python
def initialize_parameters(n_x, n_h, n_y):
    """
    参数：
        n_x - 输入层节点数量
        n_h - 隐藏层节点数量
        n_y - 输出层节点数量

    """
    # 权重参数要乘以 0.01，防止计算的输出值过大，落在激活函数梯度较小的地方
    W1 = np.random.randn(n_h, n_x) * 0.01
    b1 = np.zeros((n_h, 1))
    W2 = np.random.randn(n_y, n_h) * 0.01
    b2 = np.zeros((n_y, 1))
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters  
```

##### 初始化多层神经网络的参数
与两层网络相比，要除以/ np.sqrt(layers_dims[l - 1]) 而不是乘以 0.01，这样做的目的是为了防止（深度神经网络）出现梯度消失或梯度爆炸

```python
def initialize_parameters_deep(layers_dims):
    """
    初始化多层网络参数，包括输入层和隐藏层之间，隐藏层和输出层之间。
    参数：
        layers_dims - 包含我们网络中每层的节点数量的列表，长度为 L+1（第一个对应的是第零层-输入层的特征数）
    
    返回：
        parameters - 包含参数“W1”，“b1”，...，“WL”，“bL”的字典：
                     W1 - 权重矩阵，维度为（layers_dims [1]，layers_dims [1-1]）
                     bl - 偏向量，维度为（layers_dims [1]，1）
    """
    # 保持每回初始化的参数不同
    #np.random.seed(3)
    parameters = {}
    
    # 从第 1 层 W=(n1, n0) 到 第 L 层 W=(nL, nL-1) 的网络（包含隐藏层和输出层）参数的配置；len(layers_dims)=  L+1
    for l in range(1, len(layers_dims)):
        parameters["W" + str(l)] = np.random.randn(layers_dims[l], layers_dims[l - 1]) / np.sqrt(layers_dims[l - 1]) 
        parameters["b" + str(l)] = np.zeros((layers_dims[l], 1))
        
    return parameters
```

#### 前向传播
##### 单层网络的线性部分
LINEAR

```python
def linear_forward(Al_1, W, b):
    """
    实现单层网络前向传播的线性部分。

    参数：
        A - 来自上一层（隐藏层或输入层）的线性输出，维度为(上一层的节点数量，示例的数量）
        W - 权重矩阵，numpy数组，维度为（当前图层的节点数量，前一图层的节点数量）
        b - 偏向量，numpy向量，维度为（当前图层节点数量，1）

    返回：
         Z - 激活函数部分的输入，也称为预激活参数
         cache - 一个包含“A”，“W”和“b”的字典，存储这些变量以有效地计算后向传递
    """
    Z = np.dot(W, Al_1) + b
    assert(Z.shape == (W.shape[0], Al_1.shape[1]))
    
    # A^l-1、W^l、b^l
    linear_cache = (Al_1, W, b)
     
    return Z, linear_cache
```

##### 单层网络的激活部分
LINEAR - >ACTIVATION

```python
def linear_activation_forward(A_prev, W, b, activation):
    """
    实现 LINEAR-> ACTIVATION 这一层的前向传播，包括隐藏层的激活和输出层的激活，激活函数可选

    参数：
        A_prev - 来自上一层（隐藏层或输出层，输入层不需要激活）的线性输出，维度为(上一层的节点数量，示例数）
        W - 权重矩阵，numpy数组，维度为（当前层的节点数量，前一层的大小）
        b - 偏向量，numpy阵列，维度为（当前层的节点数量，1）
        activation - 选择在此层中使用的激活函数名，字符串类型，【"sigmoid" | "relu"】

    返回：
        A - 激活函数的输出，也称为激活后的值
        cache - 该层前向传播的cache，用于反向传播，一个包含“linear_cache：(A, W, b)”和“activation_cache：(Z)”的元组
    """
    # 区分是输出层还是隐藏层
    if activation == "sigmoid":
        Z, linear_cache = linear_forward(A_prev, W, b)
        A, activation_cache = sigmoid(Z)
    elif activation == "relu":
        Z, linear_cache = linear_forward(A_prev, W, b)
        A, activation_cache = relu(Z)
    
    assert(A.shape == (W.shape[0], A_prev.shape[1]))
    
    # 总 cache:（A^l-1、W^l、b^l，  Z^l）； A^l 不保存
    cache = (linear_cache, activation_cache)
    
    return A, cache
```

##### 多层网络的前向传播
基于单层线性输出和激活函数

```python
def L_model_forward(X, parameters):
    """
    实现输入层（X）到输出层（Yhat=AL）的计算
    
    参数：
        X - 数据，numpy数组，维度为（输入节点数量，示例数）
        parameters - initialize_parameters_deep（）的输出，储存所有神经网络层的参数
    
    返回：
        AL - 最后的激活值，预测概率（二维数组）
        caches - 缓存列表，共L-1个cache元组：
                 linear_relu_forward（）的每个cache（有L-1个，索引为从0到L-2）
                 linear_sigmoid_forward（）的cache（只有一个，索引为L-1）
    """
    caches = []
    A0 = X
    # 取整函数，参数包括 W和b对，所以要除以2，如果不整除而采用除法的话，结果是float，而range函数要求的输入必须是int
    L = len(parameters) // 2
    
    # 隐藏层（1 到 L-1层）的线性输出和激活，relu函数
    Al = A0
    for l in range(1, L):
        Al_1 = Al 
        Al, l_cache = linear_activation_forward(Al_1, parameters['W' + str(l)], parameters['b' + str(l)], "relu")
        caches.append(l_cache)
    
    # 输出层（L层）的线性输出和激活，sigmoid函数
    AL_1 = Al
    AL, L_cache = linear_activation_forward(AL_1, parameters['W' + str(L)], parameters['b' + str(L)], "sigmoid")
    caches.append(L_cache)
    
    #（1，n_examples）
    assert(AL.shape == (1, X.shape[1]))
    
    return AL, caches
```

#### 计算成本

```python
def compute_cost_2class(AL, Y):
    """
    二分类的交叉熵成本函数。

    参数：
        AL - 与标签预测相对应的概率向量，维度为（1，示例数量）
        Y - 标签向量（例如：如果不是猫，则为0，如果是猫则为1），维度为（1，数量）

    返回：
        cost - 交叉熵成本（标量）
    """
    m = Y.shape[1]
    
    # 哈达玛积的矩阵顺序不重要（element-wise product）； axis默认是None，即对所有的元素进行求和
    # print(Y.shape)
    cost = - 1/ m * np.sum( np.multiply(np.log(AL), Y) + np.multiply(np.log(1 - AL), 1 - Y) )
    # print(cost)
    
    # axis默认对所有元素求和，所以当Y是训练时，可以看作矩阵乘法，但是当Y是矩阵时就不行了
    # I = np.ones((m, 1))
    # cost1 = -1/m * np.dot( (np.log(AL) * Y + np.log(1 - AL) * (1 - Y)), I )       
    # print(np.squeeze(cost1))
    
    # 确保成本是标量
    cost = np.squeeze(cost)
    assert(cost.shape == ())

    return cost
```

#### 反向传播
##### 第 $l$ 层网络的线性部分--dW db dA
输入dZ

```python
def linear_backward(dZ, linear_cache):
    """
    为单层实现反向传播的线性部分（第L层）。基于 dZ 计算并输出 dW、db

    参数：
         dZ - 相对于（当前第l层的）线性输出的成本梯度
         cache - 来自当前层前向传播的值的元组（A_prev，W，b），即线性输出和参数

    返回：
         dA_prev - 相对于激活（前一层l-1）的成本梯度，与A_prev维度相同
         dW - 相对于W（当前层l）的成本梯度，与W的维度相同
         db - 相对于b（当前层l）的成本梯度，与b维度相同
    """
    # 用的是Wl 和 Al-1
    Al_1, W, b = linear_cache
    m = Al_1.shape[1]
    
    # 值（向量）都是一样的, 但是互为转置
    dW = np.dot(dZ, Al_1.T) / m
    # print(dW)
    # dW1 = np.dot(A_prev, dZ.T) / m
    # print(dW1)
    
    # db的结果是一致的，注意要keepdims，不要squeeze
    db = np.sum(dZ, axis=1, keepdims=True) / m    
    # print(db)
    # I = np.ones((m, 1))
    # db1 = np.dot(dZ, I) / m
    # print(db1)
    
    # 根据Wl计算dAl-1
    dAl_1 = np.dot(W.T, dZ)
    
    assert (dAl_1.shape == Al_1.shape)
    assert (dW.shape == W.shape)
    # 保存b的目的，就是在这里对照一下shape，其他地方完全用不到，但是也是有必要的！！！
    assert (db.shape == b.shape)
    
    return dAl_1, dW, db
```

##### 第 $l$ 层网络的激活部分--dZ
relu_backward 和 sigmoid_backward 来自dnn_utils.py文件

```python
def linear_activation_backward(dA, cache, activation="relu"):
    """
    实现第 l 层 LINEAR-> ACTIVATION 的反向传播。基于 dA 计算并输出 dZ
    
    参数：
         dA - 当前层l的激活后的梯度值
         cache - 该层前向传播的总cache，单个元组，值为linear_cache(A, W, b)，activation_cache（Z）
         activation - 此层中使用的激活函数名
    返回：
         dA_prev - 相对于激活（前一层l-1）的成本梯度值，与A_prev维度相同
         dW - 相对于W（当前层l）的成本梯度值，与W的维度相同
         db - 相对于b（当前层l）的成本梯度值，与b的维度相同
    """
    linear_cache, activation_cache = cache
    
    # 计算 dZ^l  dA^l-1  dW^l  db^l
    # 用不同的激活函数来区分是隐藏层还是最后的输出层
    if activation == "relu":
        dZ = relu_backward(dA, activation_cache)
        dAl_1, dW, db = linear_backward(dZ, linear_cache)
        
    elif activation == "sigmoid":
        dZ = sigmoid_backward(dA, activation_cache)
        dAl_1, dW, db = linear_backward(dZ, linear_cache)
    
    return dAl_1, dW, db
```

##### 多层模型的反向传播
基于单层网络线性部分和激活部分的反向传播；多层网络的dAL和两层网络的是一致的

也可以保存 $dZ_l$

```python
def L_model_backward(AL, Y, caches):
    """
    对[LINEAR-> RELU] *（L-1） - > LINEAR - > SIGMOID组执行反向传播，就是多层网络的向后传播
    
    参数：
     AL - 概率向量，正向传播的输出（n_L=1, n_examples）
     Y - 标签向量（例如：如果不是猫，则为0，如果是猫则为1），维度为（1，n_examples）
     caches - 包含L个cache列表：
                 linear_activation_forward（"relu"）的cache，不包含输出层
                 linear_activation_forward（"sigmoid"）的cache
    
    返回：
     grads - 具有梯度值的字典
    """
    # caches的每个元素都是每一层的所有cache组成的元组
    L = len(caches)
    m = AL.shape[1]
    
    dAL = - (np.divide(Y, AL) - np.divide(1 - Y, 1 - AL))
    # 逐元素除法，两种结果是一致的
    # dAL1 = - ( Y / AL - (1 - Y) / (1 - AL) )
      
    grads = {}
    # 也可以不加
    grads['dA' + str(L)] = dAL
    
    # 先传播输出层（层数为 L），计算 dAL-1，dZ^L-1 是中间产物，不需要输出
    cache_L = caches[L-1]
    dAL_1, dWL, dbL = linear_activation_backward(dAL, cache_L, "sigmoid")
    grads["dW" + str(L)] = dWL
    grads["db" + str(L)] = dbL    
    grads["dA" + str(L-1)] = dAL_1
    
    # 再传播隐藏层（列表的 index l 从 L-2 取到 0，层数 l 从 L-1 取到 1）
    for l in reversed(range(L - 1)):
        # 计算第 l 层的反向传播
        cache_l = caches[l]               
        # 反向传播到第 l 层，传入 dA^l（计算的是 dZ^l）和 W^l、b^l、A^l-1，输出 dA^l-1、dW^l、db^l 
        dAl = grads["dA" + str(l + 1)]
        dAl_1, dWl, dbl = linear_activation_backward(dAl, cache_l, "relu")
        grads["dW" + str(l + 1)] = dWl
        grads["db" + str(l + 1)] = dbl
        grads["dA" + str(l)] = dAl_1
    
    return grads
```

#### 参数更新

```python
def update_parameters(parameters, grads, learning_rate):
    """
    使用梯度下降更新参数
    """
    # 整除，因为每层有两个参数，W、b
    L = len(parameters) // 2 
    
    # 参数更新：从第 1 层 到 L 层（包含隐藏层和输出层）
    for l in range(L):
        parameters["W" + str(l + 1)] = parameters["W" + str(l + 1)] - learning_rate * grads["dW" + str(l + 1)]
        parameters["b" + str(l + 1)] = parameters["b" + str(l + 1)] - learning_rate * grads["db" + str(l + 1)]
        
    return parameters
```

#### 模型预测
既适合两层神经网络，也适合多层神经网络

```python
def predict(X, y, parameters):
    """
    该函数用于预测L层神经网络的结果，当然也包含两层
    
    参数：
     X - 测试集:（n0=n_features, n_examples)
     y - 标签:（nL=1, n_examples）
     parameters - 训练模型的参数
    
    返回：
     p - 给定数据集 X 的预测类别
    """
    
    m = X.shape[1]
    n = len(parameters) // 2 # 神经网络的层数
    p = np.zeros((1,m))
    
    # 前向传播，输出预测概率 probas（二维数组）
    probas, caches = L_model_forward(X, parameters)
    
    for i in range(0, probas.shape[1]):
        if probas[0, i] > 0.5:
            p[0, i] = 1
        else:
            p[0, i] = 0
    #p = np.where(probas>0.5, 1, 0)
    
    print("准确度为: "  + str(np.mean(p == y)))
        
    return p
```

#### 构建两层的神经网络
多层网络的$dA_L$和两层网络的是一致的

```python
def two_layer_model(X, Y, layers_dims, learning_rate=0.01, num_iterations=3000, print_cost=False, isPlot=True):
    """
    实现一个两层的神经网络，【LINEAR->RELU】 -> 【LINEAR->SIGMOID】
    参数：
        X - 输入的数据，维度为 (n_0=n_features，n_eaxmples)
        Y - 标签，向量，0为非猫，1为猫，维度为 (1, n_examples)
        layers_dims - 层数的向量，维度为(n_y, n_h, n_y)
        learning_rate - 学习率
        num_iterations - 迭代的次数
        print_cost - 是否打印成本值，每500次打印一次
        isPlot - 是否绘制出误差值的图谱
    返回:
        parameters - 一个包含W1，b1，W2，b2的字典变量
    """
    grads = {}
    costs = []
    
    # 初始化网络参数
    n_x, n_h, n_y = layers_dims 
    parameters = initialize_parameters(n_x, n_h, n_y)    
    W1, b1 = parameters["W1"], parameters["b1"]
    W2, b2 = parameters["W2"], parameters["b2"]   
    
    # 开始进行迭代，多迭代一次
    for i in range(num_iterations + 1):
        # 前向传播，也可以调用 L_model_forward 函数
        A1, cache1 = linear_activation_forward(X, W1, b1, "relu")
        A2, cache2 = linear_activation_forward(A1, W2, b2, "sigmoid")
        
        # 计算成本，是个标量
        cost = compute_cost_2class(A2, Y)
        
        # 反向传播，首先初始化反向传播的 dA2
        dA2 = - (np.divide(Y, A2) - np.divide(1 - Y, 1 - A2))
        
        # 反向传播，也可以调用 L_model_backward 函数
        dA1, dW2, db2 = linear_activation_backward(dA2, cache2, "sigmoid")
        dA0, dW1, db1 = linear_activation_backward(dA1, cache1, "relu")
        
        # 反向传播计算出的梯度，用于本轮的参数更新
        grads["dW1"], grads["db1"] = dW1, db1
        grads["dW2"], grads["db2"] = dW2, db2
        
        # 更新参数，修改变量 parameters
        parameters = update_parameters(parameters, grads, learning_rate)
        W1, b1 = parameters["W1"], parameters["b1"]
        W2, b2 = parameters["W2"], parameters["b2"]
        
        # 记录成本，第一次的成本值还没有进行参数更新（初始化的参数不同，损失值会不同）
        costs.append(cost)
        
        # 每迭代 500 轮打印一次成本值，第0次（参数没更新）和最后一次（num_iterations）都打印
        if i % 500 == 0 and print_cost:
            print("第", i, "次迭代，成本值为：", cost)
                
    # 迭代完成，绘制损失图
    if isPlot:
        plt.plot(costs)
        plt.ylabel('cost')
        plt.xlabel('iterations')
        plt.title("Learning rate =" + str(learning_rate))
        plt.show()
    
    return parameters
```

#### 构建多层神经网络

```python
def L_layer_model(X, Y, layers_dims, learning_rate=0.01, num_iterations=3000, print_cost=True, isPlot=True):
    """
    实现一个L层神经网络：[LINEAR-> RELU] *（L-1） - > LINEAR-> SIGMOID。
    
    参数：
        X - 输入的数据，维度为 (n_0=n_features，n_eaxmples)
        Y - 标签，向量，0为非猫，1为猫，维度为 (1, n_examples)
        layers_dims - 层数的向量，维度为(n_0, n_1,... , n_L)
        learning_rate - 学习率
        num_iterations - 迭代的次数
        print_cost - 是否打印成本值，每100次打印一次
        isPlot - 是否绘制出误差值的图谱
    
    返回：
     parameters - 模型学习的参数，可以用来预测。
    """
    
    # 初始化网络参数    
    parameters = initialize_parameters_deep(layers_dims)
        
    # 多层网络的批量迭代训练
    costs = []
    for i in range(num_iterations + 1):       
        # 前向传播
        AL, caches = L_model_forward(X, parameters)
        # 计算成本
        cost = compute_cost_2class(AL, Y)        
        # 反向传播，计算梯度
        grads = L_model_backward(AL, Y, caches) 
        # 更新参数
        parameters = update_parameters(parameters, grads, learning_rate)
        # 记录损失
        costs.append(cost)       
        # 打印成本值，第一次每训练的损失值也打印 
        if i % 500 == 0 and print_cost:
            print("第", i, "次迭代，成本值为：", cost)
                
    # 迭代完成，绘制cost变化图
    if isPlot:
        plt.plot(costs)
        plt.ylabel('cost')
        plt.xlabel('iterations')
        plt.title("Learning rate =" + str(learning_rate))
        plt.show()
        
    return parameters
```

##### 训练4层网络  
试了学习率为0.005，0.01， 0.1，都是在0.644后就超级缓慢的减小，所以不是学习率的问题  
仍是四层网络，增加神经元数，还是在0.644后就超级缓慢的减小
增加网络层数到6层, 还是在0.644后就超级缓慢的减小
所以应该不是网络复杂度不够的问题  
batch_size是全部训练集，所以应该也不存在batch_size太小的问题  
那就是数据太脏乱  http://blog.sina.com.cn/s/blog_6e32babb0102yjzq.html

问题在于前面初始化多层网络的参数时，把 / np.sqrt(layers_dims[l - 1])改为了 *0.01，导致多层网络中每层的dAL不断变小，梯度逐渐消失，从小数点后2-3位一直减小到小数点后10位（网络层数为6层时），改为/ np.sqrt(layers_dims[l - 1])后，dAL减小的速度变慢，dA0仍然保持小数点后4位 
而 / np.sqrt(layers_dims[l - 1]) 的目的是为了防止梯度消失或梯度爆炸

学习率等于0.1的话，cost会反弹，等于0.01的时候还是可以一直优化的
