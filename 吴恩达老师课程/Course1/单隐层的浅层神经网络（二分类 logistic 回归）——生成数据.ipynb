神经网络构建的一般步骤：
- 1.定义神经网络结构（输入单元的数量，隐藏单元的数量等）
 - $n_0=n_x$: 输入层的数量（特征数）
 - $n_1=n_h$: 隐藏层的数量（这里设置为4）
 - $n_2=n_y$: 输出层的数量（二分类，为1）
- 2.初始化模型的参数
- 3.循环  
 - 3.1 计算当前损失（正向传播）  
 - 3.2 计算当前梯度（反向传播）  
 - 3.3 更新参数（梯度下降）
 
前向函数对应有相应的后向函数。 因此前向传播时会 **cache** 一些值，cache的值对计算梯度很有用，在反向传播中，将使用 cache 来计算梯度

### 构建浅层逻辑回归模型

#### 初始化模型的参数
随机初始化权重矩阵（因为隐藏层的神经元数量大于1），并乘以0.01（也不是深层，可以不乘吧？？），将偏置向量初始化为零

```python
def initialize_parameters(n_x, n_h, n_y):
    """ 
    返回：
        parameters - 包含参数的字典：
            权重矩阵 W1 -（n_h，n_x）=（n1，n0）
            偏向量 b1 - （n_h，1）=（n1，1）
            权重矩阵 W2 - （n_y，n_h）=（n2，n1）
            偏向量 b2 - （n_y，1）=（n2，1）
    """
    # 指定一个随机种子，保证本文的输出与原文一致（每次初始化的数值都一样）
    np.random.seed(2) 
    W1 = np.random.randn(n_h, n_x) * 0.01
    b1 = np.zeros(shape=(n_h, 1))
    W2 = np.random.randn(n_y, n_h) * 0.01
    b2 = np.zeros(shape=(n_y, 1))
    
    parameters = {"W1" : W1,
                  "b1" : b1,
                  "W2" : W2,
                  "b2" : b2 }
    
    return parameters
```

#### 前向传播计算输出预测值
第一个激活函数使用 sigmoid() 函数（也可以使用np.tanh()函数），最后一个激活函数使用 sigmoid（只能自定义）；**cache有用的变量**

```python
def sigmoid(x):
    s = 1 / (1 + np.exp(-x))
    return s
```

```python
def forward_propagation(X, parameters):
    """
    参数：
         X - （n_x，n_examples）
         parameters - 初始化参数
    
    返回：
         A2 - 使用sigmoid()函数计算的第二次激活后的数值，即输出的预测概率向量
         cache - 包含“Z1”，“A1”，“Z2”和“A2”的字典类型变量
     """
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    
    # 前向传播计算 A2，即 Y_hat
    Z1 = np.dot(W1, X) + b1
    A1 = np.tanh(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = sigmoid(Z2)
    
    # (n_2, n_examples)
    assert(A2.shape == (1, X.shape[1]))
    
    # 保存有用的变量，反向传播时使用
    cache = {"Z1": Z1,
             "A1": A1,
             "Z2": Z2,
             "A2": A2}
    
    return (A2, cache)
```

#### 计算损失
$J=-\frac{1}{m} \sum\limits_{i=0}^{m}\Big(y^{(i)} \log \left(a^{(2,i)}\right)+\left(1-y^{(i)}\right) \log \left(1-a^{(2,i)}\right)\Big)$  ，使用的是tanh激活函数  

```python
def compute_cost(A2, Y, parameters):
    """
    计算交叉熵成本 J（标量）
    
    参数：
         parameters - 一个包含W1，B1，W2和B2的字典类型的变量
    """
    m = Y.shape[1]
    W1 = parameters["W1"]
    W2 = parameters["W2"]
    
    # 计算交叉熵成本，np.multiply是逐元素乘法（所以顺序不重要）
    logprobs = np.multiply(np.log(A2), Y) + np.multiply((1 - Y), np.log(1 - A2))
    # np.multiply是逐元素乘法，相当于 *，结果一致，已测试
    # logprobs = Y * np.log(A2) + (1 - Y) * np.log(1 - A2)
    cost = - np.sum(logprobs) / m 
    cost = float(np.squeeze(cost))
    
    # 对应二分类，所以logprobs是个行向量而非矩阵，可以使用向量乘法进行求和，已测试；如果是多分类，对应的是矩阵呢？？
    # I = np.ones((m, 1))
    # cost1 = -(1/m) * np.dot(logprobs, I)
    # cost1 = float(np.squeeze(cost1_1))
    
    assert(isinstance(cost, float)) 
    
    return cost
```

#### 反向传播计算梯度
将使用正向传播期间计算的 **cache** 来实现反向传播  

为了计算 $dZ_1$，需要计算 $g^{[1]'}(Z^{[1]})， g^{[1]}(...)$ 是tanh激活函数，所以 $g^{[1]'}(z) = 1-a^2$。需要使用 (1 - np.power(A1, 2)) 来计算 $g^{[1]'}(Z^{[1]})$。

##### 梯度计算的关系演示

```python
np.random.seed(11)
y = np.random.randn(3,2)
y
```

```python
dj_a2 = a2 * (1-a2)
dj_a2
```

```python
da2_z = -( y / a2 - (1 - y)/(1 - a2) )
da2_z
```

dz2 = a2 - y

```python
a2 - y
```

```python
dz2 = dj_a2 * da2_z
dz2
```

```python
def backward_propagation(parameters, cache, X, Y):
    """
    反向传播函数，计算参数的梯度。
    
    参数：
     parameters - 前向传播到这里的参数
     cache - 包含“Z1”，“A1”，“Z2”和“A2”的字典类型的变量。
    """
    m = X.shape[1]    
    W1 = parameters["W1"]
    W2 = parameters["W2"]
    
    A1 = cache["A1"]
    A2 = cache["A2"]
    
    # 下面的关系等式都已测试过，除了最后一层的 dA^l 可以理解外，其他的 dA 都涉及到 dZ^l 转置，还不懂 ！！！   
    # dA2 = dJ_dA2 = A2 * (1 - A2) 
    # dA2_dZ2 = - ( Y / A2 - (1 - Y) / (1 - A2) ) 
    # dZ2 = dJ_dZ2 = dJ_dA2 * dA2_dZ2      
    dZ2 = A2 - Y
    
    # dW2 = dJ_dW2 = dJ_dZ2 * dZ2_dW2，  db2 = dJ_db2 = dJ_dZ2 * dZ2_db2
    dW2 = (1 / m) * np.dot(dZ2, A1.T)    
    db2 = (1 / m) * np.sum(dZ2, axis=1, keepdims=True)
    
    # 已经测试过，两种db2的结果是一致的,shape=(1, 1)=(n_y, 1)
    # I = np.ones((m, 1))
    # db2_1 = (1 / m) * np.dot(dZ2, I)
    
    # dA1 = dJ_dZ2 * dZ2_dA1，但是 dZ2 在右边，不太懂 ！！！！
    dA1 = np.dot(W2.T, dZ2)
    # dZ1 = dJ_dA1 * dA1_dZ1，dA1 在左边，可以理解；tanh 激活，dA1_dZ1 = 1 - (A1)^2
    dZ1 = np.multiply(dA1, 1 - np.power(A1, 2))
    
    dW1 = (1 / m) * np.dot(dZ1, X.T)
    db1 = (1 / m) * np.sum(dZ1, axis=1, keepdims=True)
    
    # 已经测试过，两种db1的结果是一致的,shape=(4, 1)=(n_h, 1)
    # I = np.ones((m, 1))
    # db1_1 = (1 / m) * np.dot(dZ1, I)
    
    grads = {"dW1": dW1,
             "db1": db1,
             "dW2": dW2,
             "db2": db2 }
    
    return grads
```

#### 更新参数
需要确定一个好的学习率，成本一步步下降，而不是来回震荡

```python
def update_parameters(parameters, grads, learning_rate=1.2):
    """
    一次迭代中，梯度下降法更新参数
    """
    W1, W2 = parameters["W1"], parameters["W2"]
    b1, b2 = parameters["b1"], parameters["b2"]
    
    dW1, dW2 = grads["dW1"], grads["dW2"]
    db1, db2 = grads["db1"], grads["db2"]
    
    # 用的是减号
    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1
    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
```

#### 完整的二分类logistic浅层（单隐藏层）神经网络

```python
def nn_model(X, Y, n_h, num_iterations, print_cost=True):
    """
    参数：
        X - （n_features，n_examples）
        Y - （1，n_examples）
        n_h - 隐藏层的数量
        num_iterations - 梯度下降循环中的迭代次数
        print_cost - 如果为True，则每1000次迭代打印一次成本数值
    
    返回：
        parameters - 模型学习的参数，可以用来进行预测。
     """    
    
    # 确定网络的结构
    n_x, n_h, n_y = X.shape[0], n_h, Y.shape[0]
    
    # 初始化网络参数
    parameters = initialize_parameters(n_x, n_h, n_y)
    
    # 多次迭代更新参数
    for i in range(1, num_iterations+1):
        # 前向传播计算输出预测值和 cahe有用的变量
        A2, cache = forward_propagation(X, parameters)
        # 计算当前参数下，批量样本的成本
        cost = compute_cost(A2, Y, parameters)
        # 计算参数的梯度
        grads = backward_propagation(parameters, cache, X, Y)
        # 更新参数
        parameters = update_parameters(parameters, grads, learning_rate=0.02)
        
        # 打印前五次的损失
        if i in range(1, 6) and print_cost:
            print("第 %d 次循环，成本为：%.6f" % (i, cost))
            
        # 每迭代1000次，打印一下成本
        if print_cost and i % 1000 == 0:
            print("第 %d 次循环，成本为：%.6f" % (i, cost))
            
    return parameters
```

#### 计算预测类别的概率

```python
def predict(parameters, X):
    """
    根据学习的参数W, b来批量预测 X 的类别
    
    参数：
        X - 输入数据（n_x，m）
    
    返回
        predictions - 我们模型预测的向量（红色：0 /蓝色：1）
     
     """
    A2, cache = forward_propagation(X, parameters)
    # np.round函数：大于 0.5才返回1，小于等于0.5都返回0
    predictions = np.round(A2)
    
    return predictions
```
