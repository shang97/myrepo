#### 构建模型
神经网络构建的一般步骤：
- 1.定义神经网络结构（输入单元的数量，隐藏单元的数量等）
- 2.初始化模型的参数
- 3.循环  
3.1 计算当前损失（正向传播）  
3.2 计算当前梯度（反向传播）  
3.3 更新参数（梯度下降）

##### 定义神经网络结构
- $n_x$: 输入层的数量
- $n_h$: 隐藏层的数量（这里设置为4）
- $n_y$: 输出层的数量

```python
def layer_sizes(X , Y):
    """
    参数：
     X - 输入数据集,维度为（输入的数量，训练/测试的数量）
     Y - 标签，维度为（输出的数量，训练/测试数量）
    
    返回：
     n_x - 输入层的神经元数量
     n_h - 隐藏层的神经元数量，为4
     n_y - 输出层的神经元数量，为1
    """
    n_x = X.shape[0] # 符合吴恩达老师标准的矩阵(n_features, n_examples)
    n_h = 4 
    n_y = Y.shape[0] 
    
    return (n_x, n_h, n_y)
```

##### 初始化模型的参数
随机值初始化权重矩阵，并乘以0.01，将偏置向量初始化为零

```python
def initialize_parameters( n_x , n_h ,n_y):
    """
    参数：
        n_x - 输入层节点的数量
        n_h - 隐藏层节点的数量
        n_y - 输出层节点的数量
    
    返回：
        parameters - 包含参数的字典：
            W1 - 权重矩阵,维度为（n_h，n_x）
            b1 - 偏向量，维度为（n_h，1）
            W2 - 权重矩阵，维度为（n_y，n_h）
            b2 - 偏向量，维度为（n_y，1）

    """
    # 指定一个随机种子，保证本文的输出与原文一致。
    np.random.seed(2) 
    W1 = np.random.randn(n_h, n_x) * 0.01
    b1 = np.zeros(shape=(n_h, 1))
    W2 = np.random.randn(n_y, n_h) * 0.01
    b2 = np.zeros(shape=(n_y, 1))
    
    #使用断言确保我的数据格式是正确的
    assert(W1.shape == ( n_h , n_x ))
    assert(b1.shape == ( n_h , 1 ))
    assert(W2.shape == ( n_y , n_h ))
    assert(b2.shape == ( n_y , 1 ))
    
    parameters = {"W1" : W1,
                  "b1" : b1,
                  "W2" : W2,
                  "b2" : b2 }
    
    return parameters
```

##### 前向传播计算输出预测值
激活函数可以使用sigmoid()函数，也可以使用np.tanh()函数；**cache有用的变量**

```python
def forward_propagation(X , parameters ):
    """
    参数：
         X - 维度为（n_x，m）的输入数据。
         parameters - 初始化函数（initialize_parameters）的输出
    
    返回：
         A2 - 使用sigmoid()函数计算的第二次激活后的数值，即输出的预测概率向量
         cache - 包含“Z1”，“A1”，“Z2”和“A2”的字典类型变量
     """
    # 初始化参数
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    
    # 前向传播计算 A2，即 Y_hat
    Z1 = np.dot(W1 , X) + b1
    A1 = np.tanh(Z1)
    Z2 = np.dot(W2 , A1) + b2
    A2 = sigmoid(Z2)
    
    # 使用断言确保我的数据格式是正确的
    assert(A2.shape == (1,X.shape[1]))
    
    # 保存有用的变量，反向传播时使用
    cache = {"Z1": Z1,
             "A1": A1,
             "Z2": Z2,
             "A2": A2}
    
    return (A2, cache)
```

##### 计算损失
$J=-\frac{1}{m} \sum\limits_{i=0}^{m}\left(y^{(i)} \log \left(a^{(2,i)}\right)+\left(1-y^{(i)}\right) \log \left(1-a^{(2,i)}\right)\right)$  
使用的是tanh激活函数  

```python
def compute_cost(A2, Y, parameters):
    """
    计算方程中给出的交叉熵成本，
    
    参数：
         A2 - 使用sigmoid()函数计算的第二次激活后的数值
         Y - "True"标签向量,维度为（1，数量）
         parameters - 一个包含W1，B1，W2和B2的字典类型的变量
    
    返回：
         成本 - 交叉熵成本
    """
    
    m = Y.shape[1]
    W1 = parameters["W1"]
    W2 = parameters["W2"]
    
    # 计算交叉熵成本，np.multiply是逐元素乘法（所以顺序不重要）
    logprobs = np.multiply(np.log(A2), Y) + np.multiply((1 - Y), np.log(1 - A2))
    cost1 = - np.sum(logprobs) / m 
    cost1 = float(np.squeeze(cost1))
    
    # 对应二分类，所以logprobs是个行向量而非矩阵，可以使用向量乘法进行求和，已测试；如果是多分类，对应的是矩阵呢？？
    # I = np.ones((m, 1))
    # cost1_1 = -(1/m) * np.dot(logprobs, I)
    # cost1_1 = float(np.squeeze(cost1_1))
    
    # np.multiply是逐元素乘法，相当于 *，结果一致，已测试
    # cost2 = -(1/m) * np.sum( Y * np.log(A2) + (1 - Y) * np.log(1 - A2) )
    # cost2 = float(np.squeeze(cost2))
    # print(cost1)
    # print(cost1_1)
    # print(cost2)
    
    assert(isinstance(cost1, float))
    #assert(isinstance(cost2, float))    
    return cost1
```

##### 反向传播计算梯度
将使用正向传播期间计算的cache来实现反向传播  

为了计算 dZ1，需要计算 $g^{[1]'}(Z^{[1]})， g^{[1]}(...)$ 是tanh激活函数，所以 $g^{[1]'}(z) = 1-a^2$。需要使用 (1 - np.power(A1, 2)) 来计算 $g^{[1]'}(Z^{[1]})$。

```python
def backward_propagation(parameters, cache, X, Y):
    """
    反向传播函数，计算参数的梯度。
    
    参数：
     parameters - 包含我们的参数的一个字典类型的变量。
     cache - 包含“Z1”，“A1”，“Z2”和“A2”的字典类型的变量。
     X - 输入数据，维度为（2，数量）
     Y - “True”标签，维度为（1，数量）
    
    返回：
     grads - 包含W和b的导数一个字典类型的变量。
    """
    m = X.shape[1]
    
    W1 = parameters["W1"]
    W2 = parameters["W2"]
    
    A1 = cache["A1"]
    A2 = cache["A2"]
    
    # dA2 = - ( Y / A2 - (1 - Y) / (1 - A2) )
    # I = np.ones((m, 1))
    dZ2 = A2 - Y
    dW2 = (1 / m) * np.dot(dZ2, A1.T)
    
    # 已经测试过，两种db2的结果是一致的,shape=(1, 1)=(n_y, 1)
    db2 = (1 / m) * np.sum(dZ2, axis=1, keepdims=True)
    # I = np.ones((m, 1))
    # db2_1 = (1 / m) * np.dot(dZ2, I)
    
    dA1 = np.dot(W2.T, dZ2)
    dZ1 = np.multiply(dA1, 1 - np.power(A1, 2))
    dW1 = (1 / m) * np.dot(dZ1, X.T)
    # 已经测试过，两种db1的结果是一致的,shape=(4, 1)=(n_h, 1)
    db1 = (1 / m) * np.sum(dZ1, axis=1, keepdims=True)
    # I = np.ones((m, 1))
    # db1_1 = (1 / m) * np.dot(dZ1, I)
    
    grads = {"dW1": dW1,
             "db1": db1,
             "dW2": dW2,
             "db2": db2 }
    
    return grads
```

##### 更新参数
需要确定一个好的学习率，成本一步步下降，而不是来回震荡

```python
def update_parameters(parameters,grads,learning_rate=1.2):
    """
    一次迭代中，梯度下降法更新参数
    
    参数：
     parameters - 包含参数的字典类型的变量。
     grads - 包含导数值的字典类型的变量。
     learning_rate - 学习速率
    
    返回：
     parameters - 包含更新参数的字典类型的变量。
    """
    W1, W2 = parameters["W1"], parameters["W2"]
    b1, b2 = parameters["b1"], parameters["b2"]
    
    dW1, dW2 = grads["dW1"], grads["dW2"]
    db1, db2 = grads["db1"], grads["db2"]
    
    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1
    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
```

##### 把所有函数整合到一起
实现完整的二分类logistic单隐藏层的浅层神经网络

```python
def nn_model(X, Y, n_h, num_iterations, print_cost=True):
    """
    参数：
        X - 数据集,维度为（2，示例数）
        Y - 标签，维度为（1，示例数）
        n_h - 隐藏层的数量
        num_iterations - 梯度下降循环中的迭代次数
        print_cost - 如果为True，则每1000次迭代打印一次成本数值
    
    返回：
        parameters - 模型学习的参数，它们可以用来进行预测。
     """     
    np.random.seed(3) 
    n_x = layer_sizes(X, Y)[0]
    n_y = layer_sizes(X, Y)[2]
    
    parameters = initialize_parameters(n_x, n_h, n_y)
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    
    # 多次迭代更新参数
    for i in range(num_iterations):
        # 前向传播计算输出预测值和有用的变量
        A2, cache = forward_propagation(X, parameters)
        # 计算当前参数下，批量样本的成本
        cost = compute_cost(A2, Y, parameters)
        # 计算参数的梯度
        grads = backward_propagation(parameters, cache, X, Y)
        # 更新参数
        parameters = update_parameters(parameters, grads, learning_rate=0.02)
        
        # 每迭代1000次，打印一下成本
        if print_cost:
            if i % 1000 == 0:
                print("第 %d 次循环，成本为：%.6f" % (i, cost))
    return parameters
```

##### 计算预测类别的概率

```python
def predict(parameters, X):
    """
    根据学习的参数W, b来预测X中的每个示例的类别
    
    参数：
        parameters - 包含参数的字典类型的变量。
        X - 输入数据（n_x，m）
    
    返回
        predictions - 我们模型预测的向量（红色：0 /蓝色：1）
     
     """
    A2, cache = forward_propagation(X, parameters)
    # 大于 0.5才返回1，小于等于0.5都返回0
    predictions = np.round(A2)
    
    return predictions
```
